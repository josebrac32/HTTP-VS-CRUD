CRUD:
Es un conjunto de cuatro operaciones básicas para interactuar con datos:
 * Crear (Create): Añadir nuevos datos a un sistema.
 * Leer (Read): Recuperar o acceder a datos existentes.
 * Actualizar (Update): Modificar datos existentes.
 * Eliminar (Delete): Borrar datos existentes.

Relación:
Las operaciones CRUD se pueden implementar utilizando los métodos HTTP. Por ejemplo:
 * GET se utiliza para la operación de leer (recuperar datos).
 * POST se utiliza para la operación de crear (añadir nuevos datos).
 * PUT o PATCH se utilizan para la operación de actualizar (modificar datos).
 * DELETE se utiliza para la operación de eliminar (borrar datos).

¿Qué es una API de REST?
Una API de REST es una interfaz de programación de aplicaciones (API) que sigue los principios de diseño del estilo de la arquitectura REST. REST significa transferencia de estado representacional y consiste en un conjunto de reglas y recomendaciones para diseñar una API web.

¿Qué es una API?
Una API es un conjunto de definiciones y protocolos que se utiliza para desarrollar e integrar los sistemas de software de las aplicaciones. Suele considerarse como el contrato entre el usuario y el proveedor de información, en el cual se establece el contenido que se necesita por parte del consumidor (la llamada) y el que requiere el productor (la respuesta). Por ejemplo, el diseño de una API de servicio meteorológico podría requerir que el usuario escribiera un código postal y que el productor diera una respuesta en dos partes: la primera sería la temperatura máxima y la segunda sería la mínima.
En otras palabras, las API te permiten interactuar con una computadora o un sistema para obtener datos o ejecutar una función, de manera que el sistema comprenda la solicitud y la cumpla.
Puedes considerar la API como el mediador entre los usuarios o clientes y los recursos o servicios web que quieren obtener. Con ellas, las empresas pueden compartir recursos e información mientras conservan la seguridad, el control y la autenticación, lo cual les permite definir los accesos de cada usuario.

LEER
READ devuelve registros (o documentos o elementos) de una tabla (o colección o depósito) de una base de datos según ciertos criterios de búsqueda. La operación READ puede devolver todos los registros y algunos o todos los campos.

ACTUALIZAR
ACTUALIZAR se utiliza para modificar registros existentes en la base de datos. Por ejemplo, puede ser un cambio de dirección en una base de datos de clientes o un cambio de precio en una base de datos de productos. Al igual que LECTURA, las ACTUALIZACIONES pueden aplicarse a todos los registros o solo a algunos, según ciertos criterios.

BORRAR
Las operaciones DELETE permiten al usuario eliminar registros de la base de datos. Una eliminación definitiva elimina el registro por completo, mientras que una eliminación temporal lo marca, pero lo conserva. Por ejemplo, esto es importante en la nómina, donde es necesario conservar los registros de empleo incluso después de que un empleado deje la empresa.
HTTP (Hypertext Transfer Protocol) es un protocolo de comunicación, mientras que CRUD (Create, Read, Update, Delete) es un conjunto de operaciones básicas para la manipulación de datos en una base de datos o sistema. En otras palabras, HTTP es la forma en que los clientes y servidores se comunican, y CRUD son las acciones que se pueden realizar con los datos utilizando ese protocolo.

HTTP:
Es el protocolo fundamental para la transferencia de datos en la web. Se utiliza para solicitar recursos (como páginas web, imágenes, etc.) y recibir respuestas del servidor. Los métodos HTTP más comunes son GET (leer), POST (crear), PUT (actualizar) y DELETE (eliminar).

¿Qué es API RESTful?
La API RESTful es una interfaz que dos sistemas de computación utilizan para intercambiar información de manera segura a través de Internet. La mayoría de las aplicaciones para empresas deben comunicarse con otras aplicaciones internas o de terceros para llevar a cabo varias tareas. Por ejemplo, para generar nóminas mensuales, su sistema interno de cuentas debe compartir datos con el sistema bancario de su cliente para automatizar la facturación y comunicarse con una aplicación interna de planillas de horarios. Las API RESTful admiten este intercambio de información porque siguen estándares de comunicación de software seguros, confiables y eficientes.

CREAR
La operación CREATE añade un nuevo registro a una base de datos. En RDBMS, una fila de una tabla de base de datos se denomina registro, mientras que las columnas se denominan atributos o campos. La operación CREATE añade uno o más registros nuevos con valores de campo distintos en una tabla.
El mismo principio se aplica a las bases de datos NoSQL. Si la base de datos NoSQL está orientada a documentos, se añade un nuevo documento (por ejemplo, un documento con formato JSON con sus atributos) a la colección, lo que equivale a una tabla RDBMS. De forma similar, en bases de datos NoSQL como DynamoDB, la operación CREATE añade un elemento (equivalente a un registro) a una tabla.

Explicando CRUD
Como se mencionó, las operaciones CRUD se utilizan en aplicaciones de almacenamiento persistente, lo que significa que estas aplicaciones conservarán sus datos incluso después de que el sistema se apague. Estas son diferentes de las operaciones con datos almacenados en almacenamiento volátil, como la memoria de acceso aleatorio o los archivos de caché.
CRUD se utiliza ampliamente en aplicaciones de bases de datos. Esto incluye sistemas de gestión de bases de datos relacionales (RDBMS) como Oracle, MySQL y PostgreSQL. También incluye bases de datos NoSQL como MongoDB, Apache Cassandra y AWS DynamoDB.

¿Qué es HTTP?
El protocolo de transferencia de hipertexto (HTTP) constituye los cimientos de la red mundial, y se utiliza para cargar páginas web mediante enlaces de hipertexto. HTTP es un protocolo de capa de aplicación diseñado para transferir información entre los dispositivos conectados de la red, y se ejecuta sobre otras capas del conjunto de protocolos de la red. Un flujo típico sobre HTTP implica una máquina cliente que realiza una solicitud a un servidor, que a continuación envía un mensaje de respuesta.

¿Qué es una API REST?

Una API REST (Representational State Transfer) es un conjunto de convenciones y principios arquitectónicos para construir servicios web que permitan la comunicación entre sistemas mediante HTTP.

* Cliente-servidor: separación clara entre quien consume la API (cliente) y quien la provee (servidor).
* Sin estado (stateless): cada petición HTTP debe contener toda la información necesaria; el servidor no guarda contexto entre llamadas.
* Cacheable: las respuestas pueden ser almacenadas para mejorar rendimiento y reducir carga.
* Interfaz uniforme: uso de recursos identificados por URIs y operaciones limitadas a métodos estándar (GET, POST, PUT, DELETE).

Errores y códigos de estado HTTP
 1XX (Información): respuestas intermedias que indican que la petición sigue en proceso; no son errores.  
- 2XX (Éxito): la petición se completó correctamente; no reflejan fallos.  
- 3XX (Redirección): requieren acción adicional (nueva URL); no constituyen errores propiamente dichos.  
- 4XX (Error del cliente): la petición contiene sintaxis inválida o no está autorizada (por ejemplo, 404 No Encontrado, 403 Prohibido).  
- 5XX (Error del servidor): el servidor falló al procesar una petición válida (por ejemplo, 500 Error Interno, 503 Servicio No Disponible).
